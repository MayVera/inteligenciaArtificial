```python

# Cargar el conjunto de datos en un DataFrame
# Reemplaza 'tu_archivo.csv' con la ruta real o URL de tu conjunto de datos
df = pd.read_csv('https://raw.githubusercontent.com/Sinchiguano/chicagoUniversity/main/cta_Ridership.csv')

# Convertir la columna 'date' al formato datetime
df['date'] = pd.to_datetime(df['date'])

# Establecer la columna 'date' como índice del DataFrame
df.set_index('date', inplace=True)

# Ordenar el DataFrame por la columna de fecha ('date')
df.sort_values(by='date', inplace=True)

# Imprimir el DataFrame ordenado por fecha
print(df)
```

```python

df = pd.read_csv('https://raw.githubusercontent.com/Sinchiguano/chicagoUniversity/main/cta_Ridership.csv')

# Filtrar el DataFrame para las filas donde 'stationname' es igual a 'Garfield-South Elevated'
Garfield_South_Elevated = df[df['stationname'] == 'Garfield-South Elevated']

# Imprimir el DataFrame filtrado
print(Garfield_South_Elevated)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Establecer el tamaño de las figuras usando seaborn
sns.set(rc={'figure.figsize':(11, 4)})

# Suponiendo que Garfield_Dan_Ryan es tu DataFrame con los datos filtrados
# Reemplaza 'rides' con el nombre real de la columna en tu DataFrame
# Garfield_Dan_Ryan['rides'].plot(linewidth=0.5)

# Ejemplo de trazado de una serie temporal de rides (viajes)
plt.plot(Garfield_Dan_Ryan.index, Garfield_Dan_Ryan['rides'], linewidth=0.5)

# Personalización del gráfico
plt.title('Número de Viajes en Garfield-Dan Ryan')  # Título del gráfico
plt.xlabel('Fecha')  # Etiqueta del eje x
plt.ylabel('Número de Viajes')  # Etiqueta del eje y
plt.grid(True)  # Habilitar cuadrícula en el gráfico
plt.xticks(rotation=45)  # Rotar etiquetas del eje x para mejor visualización

# Mostrar el gráfico
plt.tight_layout()  # Ajustar diseño para evitar superposición de elementos
plt.show()
```
